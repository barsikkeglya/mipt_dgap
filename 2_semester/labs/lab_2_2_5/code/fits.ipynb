```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import numpy as np


path = r'C:\Windows\Fonts\segoeui.ttf'
path1 = r'C:\Windows\Fonts\seguisb.ttf'
path2 = r'C:\Windows\Fonts\arial.ttf'

x = ([41.5,46,24,24,22.5,30.5,35,11])
y = ([0.0327,0.03636,0.02439,0.02381,0.0198,0.025253,0.032258,0])
sqry=([])
for i in y:
    sqry.append(i**2)
print(sqry)
z = np.polyfit(x,sqry,1)
p = np.poly1d(z)

plt.figure(figsize=(6,6))

xp = np.linspace(0,50,100)
plt.xlim(0,50)
plt.ylim(0,0.04)
plt.plot(x,y,'.',color=(0,0,0),label='Experimental')
plt.plot(xp,(p(xp))**(1/2),'--',color=(0,.5,0),label='Fit with square root')

x.pop()
y.pop()

z = np.polyfit(x,y,1)
p = np.poly1d(z)
xp=np.linspace(20,50,100)
plt.plot(xp,p(xp),'--',color=(1,0,0),label='Fit using linear function')

plt.grid(linestyle='--')
plt.legend()
plt.title('Зависимость расхода $Q$ от давления $h$')
plt.xlabel('$h$, mm')
plt.ylabel('$Q$, ml/s')
plt.show()
```
